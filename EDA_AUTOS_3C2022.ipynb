<h2 style="color:#fca51f"> 2.1 Identificar como se muestran los valores faltantes y los estandarizamos (NaN) </h2>

```python
dfAutos
```

```python
dfAutos.replace("?",np.nan,inplace=True)
dfAutos.head(10)
```

<h2 style="color:#fca51f"> 2.2 Contar la cantidad de valores faltantes por variable </h2>

```python
dfValoresFaltantes = dfAutos.isnull()
dfValoresFaltantes.head(5)
```

```python
for column in dfValoresFaltantes.columns.values.tolist():
    print(column)
    print(dfValoresFaltantes[column].value_counts())
    print("")
```

```python
dfAutos.describe()
```

<h2 style="color:#fca51f"> 2.3 Analisis de imputacion de valores numericos con la media </h2>
<img src="img02.png" alt="Valores Perdidos">

```python
mean_normalized=dfAutos["peak"].astype("float").mean(axis=0) #Obteniendo la media de la variable
print("La media de normalized es: ",mean_normalized)
```

```python
dfAutos["peak"].replace(np.nan, 5100, inplace=True)
dfAutos.head(25)
```

<h2 style="color:#fca51f"> 2.4 Analisis de imputacion de valores categoricos con el mas frecuente </h2>

```python
dfAutos["doors"].value_counts().idxmax() #obtiene el dato con mayor frecuencia en la columna
```

```python
dfAutos["doors"].replace(np.nan, "four", inplace=True)
dfAutos.head(5)
```

<h2 style="color:#fca51f"> 2.5 Eliminar filas con valores faltantes </h2>

```python
dfAutos.dropna(subset = ["price"], axis=0, inplace=True)
dfAutos.reset_index(drop=True, inplace=True)
```

```python
dfValoresFaltantes = dfAutos.isnull()
dfValoresFaltantes.head(5)
for column in dfValoresFaltantes.columns.values.tolist():
    print(column)
    print(dfValoresFaltantes[column].value_counts())
    print("")
```
