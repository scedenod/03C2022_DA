<h3 style="color: #ffcd3d">3.4 Binning</h3>
<p> Se crean agrupaciones con los datos numeros enteros o punto flotantes. </p>

```python
dfTemp=dfAutos["horsepower"]
dfTemp.head(10)
```

```python
dfAutos.dtypes
```

```python
bins=np.linspace(min(dfAutos["horsepower"]), max(dfAutos["horsepower"]),4) 
#divide lo que esta en la columna, en cuatro datos especificos.
bins
```

<p> Se crearon los rangos, por ejemplo, categoria baja seria del 48 al 119, categoria media seria del 120 and 190, y categoria alta seria del 191 al 262. </p>

```python
groupNames=["Bajo","Medio","Alto"] #la asignacion es posicional y debe ser correspondiente al rango obtenido del linspace
dfAutos["horsepowerBins"]=pd.cut(dfAutos["horsepower"],bins,labels=groupNames,include_lowest=True)
#se le indica a python que divida en funcion al rango bins, con nuevo nombre groupNames
dfAutos[["horsepower","horsepowerBins"]].head(5)
```
